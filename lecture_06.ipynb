## Underfitting vs. Overfitting

![Underfitting vs. Overfitting](https://www.mathworks.com/discovery/overfitting/_jcr_content/mainParsys/image.adapt.full.medium.svg/1705396624275.svg)

**"Overfitting"** (**"ზედმეტად მორგება"**) ხდება მაშინ, როდესაც მოდელი სატრენინო მონაცემებზე ძალიან კარგ პერფორმანს გვაძლევს, თუმცა ვალიდაციის და სატესტო მონაცემებზე უჭირს მუშაობა. ეს ხდება იმიტომ, რომ მოდელი ძალიან მოერგო სატრენინგო მონაცემებს, თითქოს დაიზეპირა ისინი, და ნასწავლის განზოგადება უჭირს ახალ, უნახავ მონაცემებზე.

**"Underfitting"** (**"არასაკმარისად მორგება"**) ხდება მაშინ, როდესაც მოდელი სატრენინგო მონაცემებზე ცუდ პერფორმანს გვაძლევს, ანუ მოდელს არ შეუძლია აღბეჭდოს კავშირი მახასიათებლებსა ($x$) და სამიზნე მნიშვნელობებს ($y$) შორის.

## რეგულარიზაცია

**რეგულარიზაცია** არის ML მოდელებში "ზედმეტად მორგების" შემცირების მეთოდების ერთობლიობა. როგორც წესი, რეგულარიზაცია ერთგვარი გაცვლაა ტრენინგის სიზუსტის ზღვრულ შემცირებასა და განზოგადების (generalizability) გაზრდას შორის. რეგულარიზაციის მეთოდების გამოყენების შედეგად უმეტესად აკურატულობა სატრენინგო მონაცემებზე იკლებს, თუმცა სატესტო მონაცემებზე უფრო ზუსტ პროგნოზებს ვიღებთ.

### Bias-Variance Tradeoff

**Bias** (**მიკერძოება**) ზომავს საშუალო განსხვავებას პროგნოზირებულ და ნამდვილ მნიშვნელობებს შორის. Bias-ის მატებასთან ერთად, მოდელი სატრენინგო მონაცემებზე ნაკლებად ზუსტ პროგნოზებს იძლევა. მაღალი bias გულისხმობს ტრენინგის დროს შეცდომების მაღალ დონეს (high error).

**Variance** (**ვარიაცია/დისპერსია**) ზომავს განსხვავებას მოცემული მოდელის პროგნოზებს შორის სხვადასხვა მონაცემების გამოყენებისას. ვარიაციის მატებასთან ერთად, მოდელი ნაკლებად ზუსტად პროგნოზირებს ახალ/უნახავ მონაცემებზე. მაღალი ვარიაცია გულისხმობს შეცდომების მაღალ დონეს ტესტირებისა და ვალიდაციის დროს.

რა თქმა უნდა, მეცნიერები/ML ინჟინრები მიზნად ვისახავთ როგორც მოდელის მიკერძოების, ასევე ვარიაციის შემცირებას, თუმცა ორივეს ერთდროული შემცირება ყოველთვის არ არის შესაძლებელი და ზუსტად ამიტომ გვჭირდება რეგულარიზაცია. რეგულარიზაცია ამცირებს მოდელის ვარიაციას გაზრდილი მიკერძოების ფასად.

მიკერძოების გაზრდით და დისპერსიის შემცირებით, რეგულარიზაცია წყვეტს მოდელის "ზედმეტად მორგების" პრობლემას. "ზედმეტად მორგება" ხდება მაშინ, როდესაც სასწავლო მონაცემებში შეცდომის დონე მცირდება, ხოლო სატესტო მონაცემებში შეცდომის დონის ზრდა იწყება. სხვა სიტყვებით რომ ვთქვათ, "ზედმეტად მორგება" აღწერს მოდელებს დაბალი მიკერძოებით და მაღალი დისპერსიით. თუმცა, თუ რეგულარიზაციით შემოვიტანთ ზედმეტ მიკერძოებას, მაშინ "არასაკმარისად მორგებულ" მოდელს მივიღებთ.

მიუხედავად მისი სახელისა, "არასაკმარისად მორგება" არ ნიშნავს "ზედმეტად მორგების" საპირისპიროს. "არასაკმარისად მორგებულ" მოდელებს აქვთ მაღალი მიკერძოება და მაღალი დისპერსია. შესაბამისად, ასეთი მოდელები სატრენინგო და სატესტო მონაცემებზე მცდარ პროგნოზებს გვაძლევენ.

როგორც ვახსენე, რეგულარიზაციამ შეიძლება ასევე გამოიწვიოს მოდელის "არასაკმარისად მორგება". თუ ძალიან დიდი მიკერძოება დაინერგება რეგულარიზაციის გზით, მოდელის დისპერსიამ შეიძლება შეწყვიტოს შემცირება და ზრდა დაიწყოს. რეგულარიზაციას შეიძლება ჰქონდეს ეს ეფექტი განსაკუთრებით მარტივ მოდელებზე, ანუ მოდელებზე, რომლებსაც მინიმალური რაოდენობის პარამეტრი გააჩნიათ. განსახორციელებელი რეგულარიზაციის ტიპისა და ხარისხის განსაზღვრისას უნდა გავითვალისწინოთ მოდელის სირთულე, მონაცემები და ა.შ.

### რეგულარიზაციის მეთოდები

1. **Lasso რეგრესია** ($L_1$ რეგულარიზაცია) - რეგულარიზაციის ტექნიკა, რომელიც აჯარიმებს მაღალი მნიშვნელობის, კორელაციურ კოეფიციენტებს. მას შემოაქვს რეგულარიზაციის ცნება (ასევე უწოდებენ საჯარიმო ცნებას) მოდელის კვადრატული შეცდომების ჯამის (sum of squared errors - SSE) დანაკარგის ფუნქციაში. ეს საჯარიმო ცნება არის კოეფიციენტების/წონების აბსოლუტური მნიშვნელობების ჯამი. ის კონტროლდება ლამბდას ჰიპერპარამეტრით ($\lambda$), ის ამცირებს შერჩეული მახასიათებლების წონებს ნულამდე. ამით lasso რეგრესია მთლიანად აშორებს მოდელს მულტიკოლინეარულ მახასიათებლებს. მულტიკოლინეარულობა არის სტატისტიკური ფენომენი, რომელიც ხდება მაშინ, როდესაც რეგრესიის მოდელში ორი ან მეტი დამოუკიდებელი ცვლადი ერთმანეთთან მაღალი კორელაციაშია. სხვა სიტყვებით რომ ვთქვათ, მულტიკოლინეარულობა მიუთითებს ძლიერ წრფივ ურთიერთობაზე ცვლადებს შორის.

$$L_1\text{ regularization term} = |w_1| + |w_2| + ... + |w_n|$$

2. **Ridge რეგრესია** ($L_2$ რეგულარიზაცია) - რეგულარიზაციის ტექნიკა, რომელიც ანალოგიურად აჯარიმებს მაღალი მნიშვნელობის კოეფიციენტებს SSE დანაკარგის ფუნქციაში საჯარიმო ცნების შემოღებით. თუმცა განსხვავებით lasso რეგრესიისა, საჯარიმო ცნება არის კოეფიციენტების კვადრატების ჯამი და არა კოეფიციენტების აბსოლუტური მნიშვნელობის. Ridge რეგრესია ამცირებს მახასიათებლების წონას ნულამდე, მაგრამ არასოდეს ხდის ნულს.

$$L_2\text{ regularization term} = w_1^2 + w_2^2 + ... + w_n^2$$

3. **Elastic Net** - რეგულარიზაციის ტექნიკა, რომელიც აერთიანებს $L_1$ და $L_2$ რეგულარიზაციას დანაკარგის ფუნქციაში ჯარიმის ცნების დამატებით, რომელიც წარმოადგენს კოეფიციენტების აბსოლუტური მნიშვნელობებისა და კვადრატების ერთობლიობას. ეს იწვევს მოდელს, რომლის ზოგიერთი კოეფიციენტი ნულის ტოლია, ზოგი კი ნულთან მიახლოებულია.
4. **Early stopping** (**ადრეული შეჩერება**) - ყველაზე ხშირად გამოყენებადი რეგულარიზაციის ტექნიკაა. ის ზღუდავს გამეორებების რაოდენობას მოდელის ტრენინგის დროს. მისი მიზანია მოდელის დატრენინგება მანამ, სანამ ის არ მიაღწევს ყველაზე დაბალ სასწავლო შეცდომას, რომელიც წინ უძღვის ვალიდაციის მონაცემებზე შეცდომების პლატოს ან ზრდას.

![Early Stopping](https://www.ibm.com/content/dam/connectedassets-adobe-cms/worldwide-content/creative-assets/s-migr/ul/g/ff/70/regularization-diagrams-figure-3-model.component.complex-narrative-xl.ts=1712325946085.png/content/adobe-cms/us/en/topics/regularization/jcr:content/root/table_of_contents/body/content_section_styled/content-section-body/complex_narrative_90154187/items/content_group_1538653559/image)

რეგულარიზაციის მეთოდები შეგვიძლია განვიხილოთ ლოჯისტიკურ რეგრესიაზე. სიმარტივისთვის, წინა ლექციაზე გამოყენებულ "სათამაშო" მონაცემებს დავიხმართ დღესაც - ხელნაწერი ციფრების მონაცემებს `scikit-learn`-დან. თითოეული მონაცემი წარმოადგენს 8x8 სურათს, რომელზეც გამოსახულია ციფრი (ანუ მთელი მნიშვნელობა 0-დან 9-ის ჩათვლით). რეალურად, 8x8 მატრიცა ინახება ერთ ვექტორში 64 მახასიათებლის სახით.

````python
digits_data = load_digits()
````

რადგანაც მოდელის დატრენინგება გვსურს, აუცილებელია საწყისი მონაცემების სატრენინგო და სატესტო მონაცემებად დაშლა. ამ მაგალითში მონაცემების 80% წავა სატრენინგოში, ხოლო დარჩენილი 20% - სატესტოში.

````python
X_train, X_test, y_train, y_test = train_test_split(
    digits_data.data,
    digits_data.target,
    train_size=0.8,
    random_state=1,
    stratify=digits_data.target,
)
````

````python
print("სატრენინგო მონაცემების ზომა:", X_train.shape, y_train.shape)
print("სატესტო მონაცემების ზომა:", X_test.shape, y_test.shape)
````

თავდაპირველად შევქმნათ ლოჯისტიკური რეგრესია რეგულარიზაციის გარეშე. თუ პარამეტრებს ჩახედავთ ნახავთ, რომ ყველა `solver`-ს (მაოპტიმიზირებელ ალგორითმს) არ გააჩნია საშუალება გამოვიყენოთ რეგულარიზაციის გარეშე ან ჩვენთვის სასურველი რეგულარიზაციით:

- 'lbfgs'           -   ['l2', None]
- 'liblinear'       -   ['l1', 'l2']
- 'newton-cg'       -   ['l2', None]
- 'newton-cholesky' -   ['l2', None]
- 'sag'             -   ['l2', None]
- 'saga'            -   ['elasticnet', 'l1', 'l2', None]

ამიტომაც რეგულარიზაციის გარეშე მაგალითისთვის გამოვიყენებთ `lbfgs`-ს.

`penalty` პარამეტრი განსაზღვრავს რომელი ნორმის, რეგულარიზაციის ტექნიკის გამოყენება გვსურს. რადგანაც რეგულარიზაცია საერთოდაც არ გვინდა, ამიტომაც ვუთითებთ `None`-ს:

````python
log_reg_wo_reg = LogisticRegression(random_state=1, penalty=None, solver="lbfgs")
log_reg_wo_reg.fit(X_train, y_train)
````

$L_1$ რეგულარიზაციის გამოყენება შეგვიძლია `liblinear` ალგორითმთან ერთად, რომელიც პატარა ზომის მონაცემებზე უკეთესად მუშაობს. `C` პარამეტრი განსაზღვრავს რეგულაციის სიძლიერის შებრუნებულს, ამიტომაც რაც უფრო დაბალია, მით უფრო ძლიერია რეგულარიზაცია. აუცილებელია იყო დადებითი რეალური რიცხვი:

````python
log_reg_w_l1 = LogisticRegression(
    random_state=1, penalty="l1", solver="liblinear", C=0.02
)
log_reg_w_l1.fit(X_train, y_train)
````

როგორც $L_1$ რეგულარიზაციისას, $L_2$-ის დროსაც `C` პარამეტრის მნიშვნელობა იგივენაირად განისაზღვრება:

````python
log_reg_w_l2 = LogisticRegression(
    random_state=1, penalty="l2", solver="liblinear", C=0.02
)
log_reg_w_l2.fit(X_train, y_train)
````

ElasticNet რეგულარიზაციის მხარდაჭერა მხოლოდ `saga`-ს გააჩნია, რომელიც დიდი ზომის მონაცემებზე უფრო კარგად და სწრაფად მუშაობს ვიდრე პატარა მონაცემთა ნაკრებზე. იტერაციების მაქსიმალური რაოდენობაც შედარებით მაღალი უნდა იყოს. თანაც რადგანაც $L_1$ და $L_2$ რეგულარიზაცია ერთდროულად გამოიყენება, უნდა მივუთითოთ $L_1$ რეგულარიზაციის თანაფარდობა. `C` პარამეტრის მნიშვნელობა იგივენაირად განისაზღვრება:

````python
log_reg_w_elasticnet = LogisticRegression(
    random_state=1,
    penalty="elasticnet",
    solver="saga",
    l1_ratio=0.5,
    C=0.02,
    max_iter=2000,
)
log_reg_w_elasticnet.fit(X_train, y_train)
````

მოდი, უბრალოდ დავაკვირდეთ აკურატულობებს სატრენინგო და სატესტო მონაცემებზე. რა თქმა უნდა, ამითი არ ვიღებთ გადაწყვეტილებას თუ რომლის გამოყენება გვსურს (რომელი მოდელი იქნება ჩვენი საბოლოო გადაწყვეტილება), არამედ უბრალოდ გვინდა დავინახოთ თითოეული რეგულარიზაცია როგორ მოქმედებს. თუ მოდელების შედარება იყო ჩვენი მთავარი მიზანი და შემდგომ საბოლოო მოდელის ამორჩევა, მაშინ აუცილებლად ვალიდაციის მონაცემები უნდა გამოგვეყენებინა და მასზე დაყრდნობით აგვერჩია მოდელი.

````python
accuracies_df = pd.DataFrame(
    {
        "რეგულარიზაციის გარეშე": [
            accuracy_score(y_train, log_reg_wo_reg.predict(X_train)),
            accuracy_score(y_test, log_reg_wo_reg.predict(X_test)),
        ],
        "L1": [
            accuracy_score(y_train, log_reg_w_l1.predict(X_train)),
            accuracy_score(y_test, log_reg_w_l1.predict(X_test)),
        ],
        "L2": [
            accuracy_score(y_train, log_reg_w_l2.predict(X_train)),
            accuracy_score(y_test, log_reg_w_l2.predict(X_test)),
        ],
        "ElasticNet": [
            accuracy_score(y_train, log_reg_w_elasticnet.predict(X_train)),
            accuracy_score(y_test, log_reg_w_elasticnet.predict(X_test)),
        ],
    },
    index=["სატრენინგო", "სატესტო"],
)
````

````python
accuracies_df * 100
````

როგორც ვხედავთ, რეგულარიზაცია ნამდვილად მუშაობს და მეთოდის არჩევისას აუცილებლად უნდა მივაქციოთ ყურადღება მონაცემებს, მოდელსა და მის მახასიათებლებს და ა.შ.
